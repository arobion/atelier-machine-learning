## Entropy

$$-\sum _{ i=1 }^{ c }{ P({ x }_{ i })\log _{ b }{ P({ x }_{ i }) }  } $$

```python
import numpy as np

def calc_entropy(column):
    total = column.shape[0]
    occurences = np.bincount(column)
    entropy = 0
    for elem in occurences:
        prop = elem / total
        entropy += -(prop * np.log2(prop))
    return entropy

print("high_income", calc_entropy(data["high_income"]))
print("workclass", calc_entropy(data["workclass"] ))
```

## Information Gain

###  <span style="color:purple">Calcule le `Information Gain` de `age` en fonction de l'objectif final `high_income`</span>
 
$$Entropy(T)\quad =\sum _{ i=1 }^{ c }{ P({ x }_{ i })\log _{ b }{ P({ x }_{ i }) }  } \\ \\ IG(T,\quad A)\quad =\quad Entropy(T)-\sum _{ v\epsilon A }^{  }{ \frac { |{ T }_{ v }| }{ |T| }  } .Entropy({ T }_{ v })$$


```python
import math

#trouve le médian de age.
med = np.median(data['age'])
# crée deux subset en fonction du médian
left_split = data[data['age'] < med]
right_split = data[data['age'] >= med]

# calcule la proportion de chacuns des splits.
prop_left = left_split.shape[0] / data.shape[0]
prop_right = right_split.shape[0] / data.shape[0]
print("< median = ", prop_left)
print(">= median = ", prop_right)
```

```python
# Calcule l'entropie de high_income l'objectif final
income_entropy = calc_entropy(data["high_income"])
income_entropy
```

```python
#calculez l' `age_information_gain`
age_information_gain = income_entropy - (prop_left * calc_entropy(left_split["high_income"]) + prop_right * calc_entropy(right_split["high_income"]))
age_information_gain
```

 <span style="color:green">Attendu:0.047028661304691965</span>

 ### <span style="color:purple">Créer une fonction `calc_information_gain`</span>

```python
def calc_information_gain(data, split_name, target_name):

    # calcule l'entropy d'origine
    original_entropy = calc_entropy(data[target_name])
    # trouve le médiant de la colonne
    mean = np.mean(data[split_name])
    
    # crée deux subset en fonction du médian
    left_split = data[data[split_name] <= mean]
    right_split = data[data[split_name] > mean]
    # calcule le subset entropy de chacun des set
    to_substract = 0
    for subset in [left_split, right_split]:
        prop_subset = subset.shape[0] / data.shape[0]
        to_substract += (prop_subset * calc_entropy(subset[target_name]))    
    # Retourne information gain
    return original_entropy - to_substract

# Vérifie que la valeur coïncide avec `income`, "age", "high_income"
print(calc_information_gain(data, "age", "high_income"))
```

 <span style="color:green">Attendu: 0.0470286613047</span>

 ### <span style="color:purple">Puis une `liste information_gains` de toutes les colonnes</span>

```python
columns = ["age", "workclass", "education_num", "marital_status", "occupation", "relationship", "race", "sex", "hours_per_week", "native_country"]

information_gains = []
#faire une boucle
for col in columns:
    information_gains.append(calc_information_gain(data, col, "high_income"))
information_gains
    
```
